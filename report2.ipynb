```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans

plt.style.use("seaborn-whitegrid")
plt.rc("figure", autolayout=True)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=14,
    titlepad=10,
)

df = pd.read_csv("/Users/darren/Desktop/housing.csv")
X = df.loc[:, ["median_income", "latitude", "longitude"]]
X.head()
```

```python
kmeans = KMeans(n_clusters=6)
X["Cluster"] = kmeans.fit_predict(X)
X["Cluster"] = X["Cluster"].astype("category")
```

```python
sns.relplot(x="longitude", y="latitude", hue="Cluster", data=X, height=6)
```

```python
sns.relplot(x="median_income", y="latitude", hue="Cluster", data=X, height=6)
```

```python
sns.relplot(x="longitude", y="median_income", hue="Cluster", data=X, height=6)
```

```python
X["median_house_value"] = df["median_house_value"]
sns.catplot(x="median_house_value", y="Cluster", data=X, kind="boxen", height=6)
```
